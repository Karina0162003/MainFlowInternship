**LOAD THE DATASET**

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/Dataset.csv');
df
```

**1. Filtering data based on conditions**

```python
#Filter rows where Y-Kappa > 22
yKappaFilter = df[df['Y-Kappa'] > 22]
print(yKappaFilter)

#Filter rows where Y-Kappa < 22
yKappaFilters = df[df['Y-Kappa'] < 22]
print(yKappaFilters)
```

```python
# Filter rows where BlowFlow is between 1100 and 1400
blowFlowFilter = df[(df['BlowFlow'] > 1100) & (df['BlowFlow'] < 1400)]
print(blowFlowFilter)

# Filter rows where BlowFlow is between 1100 and 1400
blowFlowFilter = df[(df['BlowFlow'] > 1100) | (df['BlowFlow'] < 1400)]
print(blowFlowFilter)
```

```python
# Filter rows where ChipLevel4 is greater than its mean
chip_level_mean = df['ChipLevel4 '].mean()
filterChip = df[df['ChipLevel4 '] > chip_level_mean]
print(filterChip)
```

```python
# Filter rows where T-upperExt-2 is less than or equal to 350 and T-lowerExt-2 is greater than 329
filterData = df[(df['T-upperExt-2 '] <= 350) & (df['T-lowerExt-2  '] > 329)]
print(filterData)
```

```python
#Filter rows where ChipRate is exactly 16.52
filterChipRate = df[(df['ChipRate'] >= 16.52) & (df['ChipRate'] <= 19.50)]
print(filterChipRate)
```

```python
#Filter rows where Observation starts with "31-01"
filterObservation = df[df['Observation'].str.startswith('31-01')]
print(filterObservation)
```

```python
# Filter rows where UCZAA is greater than the median value of UCZAA
median_uczaa = df['UCZAA'].median()
filterUczaa = df[df['UCZAA'] > median_uczaa]
print(filterUczaa)
```

```python
# Filter rows where BlackFlow-2 is not null and AA-Wood-4 is less than 45
filterData = df[df['BlackFlow-2 '].notnull() & (df['AA-Wood-4  '] < 45)]
print(filterData)
```

```python
# Filter rows where T-Top-Chips-4 or SulphidityL-4 is greater than 250
filterDatas = df[(df['T-Top-Chips-4 '] > 250) | (df['SulphidityL-4 '] > 250)]
print(filterDatas)
```

```python
# Filter rows where AA-Wood-4 is between 40 and 50
filterWood = df[(df['AA-Wood-4  '] >= 10) & (df['AA-Wood-4  '] <= 20)]
print(filterWood)
```

**Handling Missing Value**

```python
# Counting Null value from each column
filtered_df = df.isnull().sum()
print(filtered_df)
```

```python
# Fill missing values with the mean for multiple columns
columns_to_fill = ['ChipRate','BF-CMratio','BlowFlow','ChipLevel4 ','T-upperExt-2 ',
'T-lowerExt-2  ','UCZAA','WhiteFlow-4 ','AAWhiteSt-4 ','AA-Wood-4  ','ChipMoisture-4 ',
'SteamFlow-4 ','Lower-HeatT-3','Upper-HeatT-3 ','ChipMass-4 ','WeakLiquorF ',
'BlackFlow-2 ','WeakWashF ','SteamHeatF-3 ','T-Top-Chips-4 ','SulphidityL-4 ']

for column in columns_to_fill:
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)

df.isnull().sum()
```

**Calculating summary statistics.**

```python
# Method 2: Individual summary statistics
df1 = df.drop(columns=['Observation'])
mean_values = df1.mean()  # Mean
median_values = df1.median()  # Median
std_deviation = df1.std()  # Standard Deviation
min_values = df1.min()  # Minimum
max_values = df1.max()  # Maximum
count_values = df1.count()  # Count of non-null values

# Display individual summary statistics
print("Mean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nStandard Deviation:\n", std_deviation)
print("\nMinimum Values:\n", min_values)
print("\nMaximum Values:\n", max_values)
print("\nCount of Non-null Values:\n", count_values)
```

```python
# Method 4: Additional summary statistics
unique_values = df1.nunique()  # Number of unique values per column
skewness = df1.skew()  # Skewness of each numerical column
kurtosis = df1.kurtosis()  # Kurtosis of each numerical column

# Display additional summary statistics
print("Number of Unique Values:\n", unique_values)
print("\nSkewness:\n", skewness)
print("\nKurtosis:\n", kurtosis)
```
