```python
import pandas as pd
```

```python
data = pd.read_csv("/content/householdtask3.csv")
data
```

```python
import matplotlib.pyplot as plt
```

```python
years = data['year']
total_households = data['tot_hhs']
own_wm_prop = data['own_wm_prop']
```

```python
plt.figure(figsize=(12, 5))

# Bar chart
plt.subplot(1, 2, 1)
plt.bar(years, total_households, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.title('Total Households by Year')
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
plt.plot(years, own_wm_prop, marker='o', linestyle='-', color='green')
plt.xlabel('Year')
plt.ylabel('Proportion (%)')
plt.title('Proportion of Households Owning Washing Machines by Year')
plt.xticks(rotation=45)

# Show the plots
plt.tight_layout()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['year'], data['income'], marker='o', linestyle='-', color='green', label='Average Income')
plt.plot(data['year'], data['expenditure'], marker='o', linestyle='-', color='red', label='Average Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Average Income and Expenditure by Year')
plt.legend()
plt.show()
```
